```python
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
x = np.linspace(-10,10,201)
y = np.linspace(-10,10,201)
X,Y = np.meshgrid(x,y)
```

```python
z_bath = X+Y
```

```python
z_tnm = np.zeros((10,201,201),dtype=float)

for t in range(10):
    tsunami = t*np.sin(X+t/10*np.pi)
    z_tnm[t] = tsunami
    
```

```python
fig = go.FigureWidget()
fig.add_surface(z=z_bath)

for i in range(10):
    fig.add_surface(z=z_tnm[i])

fig
```
